```python
#Ans2: A scatter plot is a type of data visualization that displays values for two variables as points on a Cartesian plane
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='blue', marker='o')
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.grid(True)
```

```python
#Ans3:The subplot() function in Matplotlib is used to create multiple subplots within a single figure.
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x1, y1, marker='o', color='blue')
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2, 2, 2)
plt.plot(x2, y2, marker='o', color='green')
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2, 2, 3)
plt.plot(x3, y3, marker='o', color='red')
plt.title('Line Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2, 2, 4)
plt.plot(x4, y4, marker='o', color='purple')
plt.title('Line Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()

plt.show()
```

```python
#Ans4:A bar plot (or bar chart) is a type of chart that represents categorical data with rectangular bars. The length of each bar is proportional to the value it represents.
#Uses-
#a.Comparison
#b. Categorical Data
#c. Simplicity
#d. Trend Analysis
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.bar(company, profit, color='blue')
plt.title('Vertical Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.xticks(rotation=45)  
plt.subplot(1, 2, 2)
plt.barh(company, profit, color='pink')
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')

plt.tight_layout()

plt.show()
```

```python
#Ans5:A box plot (or box-and-whisker plot) is a graphical representation of the distribution of a dataset. It provides a summary of key statistical measures, including the median, quartiles, and potential outliers.
#Uses-
#a.Distribution Summary
#b.Comparison
#c.Outliers Detection
#d.Visual Insight
np.random.seed(0)  # For reproducibility
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], tick_labels=['Box 1', 'Box 2'])

# Add title and labels
plt.title('Box Plot of Two Datasets')
plt.xlabel('Dataset')
plt.ylabel('Value')

# Show plot
plt.grid(True)  # Optional: Add grid for better readability
plt.show()
```
